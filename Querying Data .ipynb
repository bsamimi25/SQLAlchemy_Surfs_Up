```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine,inspect, func, desc

%matplotlib notebook
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import datetime as dt

# Data Setup
engine=create_engine("sqlite:///Resources/hawaii.sqlite")

# refelct an existing database into a new model 
Base= automap_base()

# reflect the tables
Base.prepare(engine, reflect= True)
```

```python
# Save References to each Table
Measurement= Base.classes.measurement
Station= Base.classes.station
```

```python
# Create a Session link from Python to Database
session= Session(engine)
```

```python
# Query date and Precipitation
results= session.query(Measurement.date, Measurement.prcp).all()
```

```python
# Print out Results
for result in results:
    print(result)
```

# Climate Analysis of Oahu
## Precipitation Analysis

```python
# Find the Date One Year Ago to begin Analysis
# This function allows us to trace back a certain number of days 
prev_year= dt.date(2017,8,23)-dt.timedelta(365)
```

```python
print("Date range is starting from 2016-08-23 and ending at 2017-08-23")
```

```python
# Get Precipitation Scores
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# Save Results into a dataframe 
df= pd.DataFrame(results, columns=['date', 'precipitation'])
```

```python
# Do not want to make a new DataFrame by using inplace=True 
df.set_index(df['date'], inplace=True)
print(df)
```

```python
# Need to delete the second date column, going to set all columns to strings except index 
print(df.to_string(index=False))
```

```python
# Sort the Data
df=df.sort_index()
print(df.to_string(index=False))
```

```python
# Plot the DataFrame
df.plot()

#Edit some details of graph 
plt.ylabel=('Precipitation')
plt.xlabel=("Date")
plt.xticks(rotation='45')
plt.tight_layout()
plt.show()
```

```python
# Summary Statistics
df.describe()
```

## Station Analysis

```python
# Query to determine the number of Stations 
session.query(func.count(Station.station)).all()
```

```python
# Query to determine most active stations
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).all()
```

# Temperature Analysis

```python
# Most active station and temperature scores
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == 'USC00519281').all()
```

```python
# Filtering out all other stations and filtering the dates

results= session.query(Measurement.tobs).\
filter(Measurement.station== 'USC00519281').\
filter(Measurement.date>= prev_year).all()

print(results)
```

```python
# Need to put into a dataframe 
results= pd.DataFrame(results, columns=['tobs'])
print(results)
```
